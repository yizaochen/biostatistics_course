```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.stats import norm
```

<h2 style="color:blue">Q1: Use Z table and find the values of $p=\text{Pr}(Z>k)$</h2>

<h4 style="color:blue">(1) If $\text{Pr}(Z>k)=0.2$, $k=$?</h4>

```python
z1 = 0.845
z2 = 4
area = 1 - norm(0, 1).cdf(z1)
x = np.arange(z1, z2, 0.001) # range of x in spec
x_all = np.arange(-10, 10, 0.001) # entire range of x, both in and out of spec
y = norm.pdf(x,0,1)
y2 = norm.pdf(x_all,0,1)

fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('fivethirtyeight')
ax.plot(x_all,y2)

ax.fill_between(x,y,0, alpha=0.3, color='b')
ax.fill_between(x_all,y2,0, alpha=0.1)
ax.set_xlim([-4,4])
#ax.set_xlabel('# of Standard Deviations Outside the Mean')
ax.set_xticks([-4, -3, -2, -1, 0, 1, 1.85, 3, 4])
ax.set_yticklabels([])
ax.set_title(f'P(z > {z1:.3f})={area:.2f}')
plt.tight_layout()
#plt.savefig('Q7_1_1.png', dpi=200)
plt.show()
```

<h2 style="color:blue">Q2</h2>

```python
n = 14
mu = 86
xbar = 91.1
sigma = 8
sigma_div_sqrt_n = sigma / np.sqrt(n)
interval = 0.1

xticks = list()
for idx in [-4, -3, -1.96, -1, 0, 1, 1.96, 3, 4]:
    xticks.append(mu + idx * sigma_div_sqrt_n)
```

```python
x_min = mu - 4 * sigma_div_sqrt_n
x_max = mu + 4 * sigma_div_sqrt_n
z1 = mu - 1.96 * sigma_div_sqrt_n
z2 = mu + 1.96 * sigma_div_sqrt_n
area1 = norm(mu, sigma_div_sqrt_n).cdf(z1)
area2 = norm(mu, sigma_div_sqrt_n).cdf(z2)
area3 = area2 - area1
area_total = norm(mu, sigma_div_sqrt_n).cdf(x_max)
p = area3 / area_total

interval_xmin = xbar - 1.96 * sigma_div_sqrt_n
interval_xmax = xbar + 1.96 * sigma_div_sqrt_n
interval_x = np.arange(interval_xmin, interval_xmax, interval)
interval_y = [-0.02] * len(interval_x)


x = np.arange(z1, z2+interval, interval) # range of x in spec
x_all = np.arange(x_min, x_max, interval) # entire range of x, both in and out of spec
y = norm.pdf(x, mu, sigma_div_sqrt_n)
y2 = norm.pdf(x_all, mu, sigma_div_sqrt_n)

fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('fivethirtyeight')
ax.plot(x_all,y2)
ax.axvline(mu, color='red', alpha=0.2)
ax.fill_between(x,y,0, alpha=0.3, color='b')
ax.fill_between(x_all,y2,0, alpha=0.1)
ax.set_xlabel(r'$\bar{x}$')
ax.set_xticks(xticks)
ax.set_yticklabels([])
title = f'P({z1:.2f} $\leq$' + r'$\bar{x} \leq$' + f'{z2:.2f})={p:.2f}'
ax.set_title(title)

ax.scatter(xbar,-0.02, s=100, color='black', label=)
ax.plot(interval_x, interval_y, color='green', label="Confidnece Interval")
ax.legend()

plt.tight_layout()
#plt.savefig('Q7_2_2.png', dpi=200)
plt.show()
```

```python
x_min = mu - 4 * sigma_div_sqrt_n
x_max = mu + 4 * sigma_div_sqrt_n
z1 = mu - 1.645 * sigma_div_sqrt_n
z2 = mu + 1.645 * sigma_div_sqrt_n
area1 = norm(mu, sigma_div_sqrt_n).cdf(z1)
area2 = norm(mu, sigma_div_sqrt_n).cdf(z2)
area3 = area2 - area1
area_total = norm(mu, sigma_div_sqrt_n).cdf(x_max)
p = area3 / area_total

x = list(np.arange(z1, z2, interval)) + [z2] # range of x in spec
x_all = np.arange(x_min, x_max, interval) # entire range of x, both in and out of spec
y = norm.pdf(x, mu, sigma_div_sqrt_n)
y2 = norm.pdf(x_all, mu, sigma_div_sqrt_n)

fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('fivethirtyeight')
ax.plot(x_all,y2)
ax.axvline(mu, color='red', alpha=0.2)

ax.plot(interval_x, interval_y, color='green')


ax.fill_between(x,y,0, alpha=0.3, color='b')
ax.fill_between(x_all,y2,0, alpha=0.1)
ax.set_xlabel(r'$\bar{x}$')
ax.set_xticks(xticks)
ax.set_yticklabels([])
title = f'P({z1:.2f} $\leq$' + r'$\bar{x} \leq$' + f'{z2:.2f})={p:.2f}'
ax.set_title(title)
plt.tight_layout()
#plt.savefig('Q7_2_3.png', dpi=200)
plt.show()
```

```python
n = 14
mu = 86
xbar = 91.1
sigma = 8
sigma_div_sqrt_n = sigma / np.sqrt(n)
interval = 0.1

xticks = list()
for idx in range(-4,5):
    xticks.append(mu + idx * sigma_div_sqrt_n)
```

```python
x_min = mu - 4 * sigma_div_sqrt_n
x_max = mu + 4 * sigma_div_sqrt_n
x_all = np.arange(x_min, x_max, interval) # entire range of x, both in and out of spec
y_all = norm.pdf(x_all, mu, sigma_div_sqrt_n)
```

```python
# 95% Confidence Interval
critival_value = 1.96
c95_y = -0.02
c95_xmin = xbar - critival_value * sigma_div_sqrt_n
c95_xmax = xbar + critival_value * sigma_div_sqrt_n
c95_xlist = np.arange(c95_xmin, c95_xmax, interval)
c95_ylist = [c95_y] * len(c95_xlist)
```

```python
# 90% Confidence Interval
critival_value = 1.645
c90_y = -0.03
c90_xmin = xbar - critival_value * sigma_div_sqrt_n
c90_xmax = xbar + critival_value * sigma_div_sqrt_n
c90_xlist = np.arange(c90_xmin, c90_xmax, interval)
c90_ylist = [c90_y] * len(c90_xlist)
```

```python
plt.style.use('fivethirtyeight')
```

```python
fig, ax = plt.subplots(figsize=(9,6))

ax.plot(x_all,y_all, color="blue", label="Sampling Distribution",alpha=0.5)
ax.fill_between(x_all,y_all,0, color="blue", alpha=0.1)
ax.set_yticklabels([])
ax.axvline(mu, color='red', label=r'$\mu$', alpha=0.5)
ax.set_xticks(xticks)

ax.plot(c95_xlist, c95_ylist, color='green', label="95% Confidence Interval")
ax.scatter(xbar, c95_y, s=100, color='green')

ax.plot(c90_xlist, c90_ylist, color='magenta', label="90% Confidence Interval")
ax.scatter(xbar, c90_y, s=100, color='magenta')

ax.legend()
plt.tight_layout()
#plt.savefig('HW7_Q3.png', dpi=200)
plt.show()
```
