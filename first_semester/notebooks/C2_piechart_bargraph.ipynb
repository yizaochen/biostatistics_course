```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2011 year, from the National Center for Health Statistics
causes = ['Heat disease', 'Cancer', 'Chronic respiratory disease', 
          'Cerebrovascular disease', 'Accidental death', 'Alzheimerâ€™s disease',
          'Diabetes', 'Flu and pneumonia', 'Kidney disorder', 'Suicide']
counts = [596339, 575313, 143382, 128931, 122777, 84691, 73282, 53667, 45731, 38285]
```

### Calculate percentage

```python
counts = np.array(counts)
total_counts = counts.sum()
percentage = list()
for count in counts:
    percentage.append(count / total_counts)
percentage = np.array(percentage) * 100 # transform to %
```

```python
percentage
```

### Pie chart

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(8,4))
ax.pie(percentage, labels=causes, autopct='%1.0f%%')
plt.show()
```

### Bar plots

```python
w = 0.5
n_causes = len(causes)
x = range(n_causes)

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(8,4))
ax.bar(x, percentage, w)
ax.set_xticks(x)
ax.set_xticklabels(causes, rotation=90)
ax.set_xlabel('Top 10 causes of death', fontsize=12)
ax.set_ylabel('Percent of top ten', fontsize=12)
plt.show()
```
