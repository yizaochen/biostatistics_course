```python
from scipy.stats import t
import matplotlib.pyplot as plt
import numpy as np
```

<h2 style="color:blue">Q1</h2>

```python
df = 15
xvalue = -1.753
xticks = [xvalue, -1, 0, 1, 2]
x_fill = np.linspace(t.ppf(0.01, df), xvalue, 100)
y_fill = t.pdf(x_fill, df)
```

```python
fig, ax = plt.subplots(1, 1)
area = t.cdf(xvalue, df)
x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
ax.plot(x, t.pdf(x, df), 'b-', lw=2, alpha=0.6, label=f't(df={df})')
ax.axvline(xvalue, color='r', lw=2, alpha=0.5)
ax.axhline(0, color='black')
title = f'P(X<{xvalue})={area:.2f}'
ax.set_title(title)
ax.set_xticks(xticks)

ax.fill_between(x_fill, y_fill, 0, alpha=0.3, color='b')

ax.set_ylim(0, 0.42)
plt.tight_layout()
#plt.savefig('HW8_Q1.png', dpi=150)
plt.show()
```

<h2 style="color:blue">Q2</h2>

```python
df = 8
xvalue = 2.306
xticks = [-xvalue, -1, 0, 1, xvalue]
x_fill_1 = np.linspace(t.ppf(0.01, df), -xvalue, 100)
y_fill_1 = t.pdf(x_fill_1, df)
x_fill_2 = np.linspace(xvalue, t.ppf(0.99, df), 100)
y_fill_2 = t.pdf(x_fill_2, df)
```

```python
fig, ax = plt.subplots(1, 1)
area = 1 - (t.cdf(xvalue, df) - t.cdf(-xvalue, df))
x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
ax.plot(x, t.pdf(x, df), 'b-', lw=2, alpha=0.6, label=f't(df={df})')
ax.axvline(-xvalue, color='r', lw=2, alpha=0.5)
ax.axvline(xvalue, color='r', lw=2, alpha=0.5)
ax.axhline(0, color='black')
title = f'P(X<{-xvalue})+P(X>{xvalue})={area:.2f}'
ax.set_title(title)
ax.set_xticks(xticks)

ax.fill_between(x_fill_1, y_fill_1, 0, alpha=0.3, color='b')
ax.fill_between(x_fill_2, y_fill_2, 0, alpha=0.3, color='b')

ax.set_ylim(0, 0.42)
plt.tight_layout()
#plt.savefig('HW8_Q2.png', dpi=150)
plt.show()
```

<h2 style="color:blue">Q3</h2>

```python
data = [0.95, 1.02, 1.01, 0.98]
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)
print(f'Sample mean is {mean:.4f}')
print(f'STD is {std:.4f}')
```

```python
t_critical = 3.182
left = mean - t_critical * std/np.sqrt(n)
right = mean + t_critical * std/np.sqrt(n)
print(f'The interval is from {left:.4f} to {right:.4f}')
```

```python
t_sample = (mean-1)/(std/np.sqrt(n))
print(f't-sample is {t_sample:.4f}')
```

```python
df = 3
xvalue = 3.182
xticks = [-xvalue, -2, -1, 0, 1, 2, xvalue]
x_fill_1 = np.linspace(t.ppf(0.01, df), -xvalue, 100)
y_fill_1 = t.pdf(x_fill_1, df)
x_fill_2 = np.linspace(xvalue, t.ppf(0.99, df), 100)
y_fill_2 = t.pdf(x_fill_2, df)
```

```python
fig, ax = plt.subplots(1, 1)
area = 1 - (t.cdf(xvalue, df) - t.cdf(-xvalue, df))
x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
ax.plot(x, t.pdf(x, df), 'b-', lw=2, alpha=0.6, label=f't(df={df})')
ax.axvline(-xvalue, color='r', lw=2, alpha=0.5)
ax.axvline(xvalue, color='r', lw=2, alpha=0.5)
ax.axvline(t_sample, color='green', alpha=0.7, label=r'$t_{obs}=$'+f'{t_sample:.3f}')
ax.axhline(0, color='black')
title = f'P(X<{-xvalue})+P(X>{xvalue})={area:.2f}'
ax.set_title(title)
ax.set_xticks(xticks)

ax.fill_between(x_fill_1, y_fill_1, 0, alpha=0.3, color='b')
ax.fill_between(x_fill_2, y_fill_2, 0, alpha=0.3, color='b')
ax.legend()
ax.set_ylim(0, 0.42)
plt.tight_layout()
#plt.savefig('HW8_Q3.png', dpi=150)
plt.show()
```
