```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.stats import norm, binom
```

<h2 style="color:blue">Q2</h2>

```python
mu = 123.
sigma = 0.08
n = 3
sigma_bar = 0.08 / np.sqrt(3)

x = np.arange(122, 124, 0.01)
y_population = norm.pdf(x, mu, sigma)
y_sample = norm.pdf(x, mu, sigma_bar)
```

```python
fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('fivethirtyeight')
ax.plot(x, y_population, color='blue', lw=4, label='Population', alpha=0.4)
ax.plot(x, y_sample, color='red', lw=4, label='Sample Mean(n=3)', alpha=0.4)
ax.legend()
ax.set_yticklabels([])
plt.tight_layout()
#plt.savefig('Q2_a.png', dpi=200)
plt.show()
```

```python
z1 = -4.
z2 = 1.85
x = np.arange(z1, z2, 0.001) # range of x in spec
x_all = np.arange(-10, 10, 0.001) # entire range of x, both in and out of spec
y = norm.pdf(x,0,1)
y2 = norm.pdf(x_all,0,1)
```

```python
fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('fivethirtyeight')
ax.plot(x_all,y2)

ax.fill_between(x,y,0, alpha=0.3, color='b')
ax.fill_between(x_all,y2,0, alpha=0.1)
ax.set_xlim([-4,4])
#ax.set_xlabel('# of Standard Deviations Outside the Mean')
ax.set_xticks([-4, -3, -2, -1, 0, 1, 1.85, 3, 4])
ax.set_yticklabels([])
ax.set_title('z < 1.85')
plt.tight_layout()
#plt.savefig('Q5_a.png', dpi=200)
plt.show()
```

<h2 style="color:blue">Q3</h2>

```python
mu = 0.9
sigma = 0.15
n = 125
sigma_bar = sigma / np.sqrt(n)

x = np.arange(mu-4*sigma, mu+4*sigma, 0.001)
y_population = norm.pdf(x, mu, sigma)
y_sample = norm.pdf(x, mu, sigma_bar)
```

```python
fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('fivethirtyeight')
#ax.plot(x, y_population, color='blue', lw=4, label='Population', alpha=0.4)
ax.plot(x, y_sample, color='red', lw=4, label='Sample Mean(n=125)', alpha=0.4)
ax.legend()
ax.set_yticklabels([])
ax.set_xticks([mu-3*sigma_bar, mu-2*sigma_bar, mu-1*sigma_bar, mu, mu+1*sigma_bar, mu+2*sigma_bar, mu+3*sigma_bar])
ax.set_xlim(mu-4*sigma_bar, mu+4*sigma_bar)
plt.tight_layout()
plt.savefig('Q3_a.png', dpi=200)
plt.show()
```

<h2 style="color:blue">Q5</h2>

```python
n = 15
p = 0.2
mu = n * p
sigma = np.sqrt(n*p*(1-p))
```

```python
x = range(0, n+1)
y = binom.pmf(x, n, p)
fig, axes = plt.subplots(ncols=2, figsize=(12,4))

ax = axes[0]
ax.plot(x, y, 'bo', ms=8)
ax.vlines(x, 0, y, colors='b', lw=5, alpha=0.5)
ax.axvline(mu, color='red', label='$\mu=np$')
ax.set_xticks(x)
ax.set_ylabel("P(X=x)")
ax.set_xlabel("x")
ax.legend()
ax.set_title("$Binomial(n=15,p=0.2)$")

ax = axes[1]
x = np.arange(mu-3*sigma, mu+3*sigma, 0.01)
y = norm.pdf(x, mu, sigma)
ax.plot(x,y)
ax.set_xticks(range(0, n+1))
ax.set_ylabel("P(X=x)")
ax.set_xlabel("x")
ax.set_title("$N(3, 1.549)$")

plt.show()
```

<h2 style="color:blue">Q6-(a)</h2>

```python
n = 100
p = 0.75
mu = n * p
sigma = np.sqrt(n*p*(1-p))
```

```python
x = range(0, n+1)
y = binom.pmf(x, n, p)
fig, axes = plt.subplots(ncols=2, figsize=(12,4))

ax = axes[0]
#ax.plot(x, y, 'bo', ms=8)
ax.vlines(x, 0, y, colors='b', lw=5, alpha=0.5)
ax.axvline(mu, color='red', label='$\mu=np$')
ax.set_xticks(x[::10])
ax.set_ylabel("P(X=x)")
ax.set_xlabel("x")
ax.legend()
ax.set_title(f'Binomial(n={n},p={p:.2f})')

ax = axes[1]
x = np.arange(mu-3*sigma, mu+3*sigma, 0.01)
y = norm.pdf(x, mu, sigma)
ax.plot(x,y)
ax.set_xticks(range(0, n+1)[::10])
ax.set_ylabel("P(Y=y)")
ax.set_xlabel("y")
ax.set_title(f'N({mu:.0f}, {sigma:.2f})')

#plt.savefig(f'Q6_a.png', dpi=200)
plt.show()
```

```python
def f(x):
    n = 100
    p = 0.75
    mu = n * p
    sigma = np.sqrt(n*p*(1-p))
    return norm.pdf(x, mu, sigma)
```
